```python
# 2. 랜선자르기
# 40점. 마지막3개 타임리밋
import sys
#sys.stdin = open('input.txt','rt')
k,n = map(int,input().split())
list1 = [int(input()) for _ in range(k)]
ans = sum(list1) // n
for i in range(ans,0,-1) : 
    a = sum([x//i for x in list1])
    if a == n : 
        break
print(i)

# 복습풀이
import sys
#sys.stdin = open('input.txt','rt')
def Count(len) :
    cnt = 0
    for x in a : 
        cnt += x//len
    return cnt
n,k = map(int, input().split())
a = [int(input()) for _ in range(n)]
lt = 1
rt = max(a)
while lt <= rt: 
    mid = (lt+rt)//2
    if Count(mid) >= k : 
        res = mid
        lt = mid+1
    else : 
        rt = mid-1
print(res)
```

```python
# 4. 마구간 정하기
# 20점. 2,3,4,5 타임리밋
import sys
#sys.stdin = open('input.txt','rt')
n,c = map(int,input().split())
list1 = [int(input()) for _ in range(n)]
list1.sort()
import itertools
list2 = list(itertools.combinations(list1,c))
def distmaker(alist) : 
    ans = float('inf')
    for i in range(1,len(alist)) : 
        dist = alist[i] - alist[i-1]
        if dist<ans :
            ans = dist
    return ans
list3 = list(map(distmaker,list2))
print(max(list3))

# 복습풀이
import sys
#sys.stdin = open('input.txt','rt')
def Count(len) :
    cnt = 1
    ep = a[0] 
    for i in a[1:] : 
        if i-ep >= len : 
            cnt += 1
            ep = i
    return cnt
n,c = map(int, input().split())
a = [int(input()) for _ in range(n)]
a.sort()
lt = 1
rt = a[n-1]
while lt <= rt: 
    mid = (lt+rt)//2
    if Count(mid) >= c : 
        res = mid
        lt = mid+1
    else : 
        rt = mid-1
print(res)
```

```python
# 5. 회의실 배정
# 복습풀이
import sys
#sys.stdin = open('input.txt','rt')
n = int(input())
a = [list(map(int,input().split())) for _ in range(n)]
a.sort(key = lambda x : [x[1],x[0]])
et=a[0][1]
cnt=1
for x,y in a :
    if x >= et : 
        cnt += 1
        et = y
print(cnt)
```

```python
# 6. 씨름선수
import sys
#sys.stdin = open('input.txt','rt')
n = int(input())
listh=[]
listw=[]
for _ in range(n) : 
    a,b = map(int,input().split())
    listh.append(a)
    listw.append(b)

cnt=n
for i in range(n) : 
    listhcheck = [idx for idx, x in enumerate(listh) if x>listh[i]]
    if listhcheck == [] : #전체 중 키가 제일 크다
        continue
    listwcheck = [idx for idx, x in enumerate(listw) if x>listw[i]]
    if listwcheck == [] : #전체 중 몸무게가 제일 크다
        continue
    checklist = [x for x in listhcheck if x in listwcheck]
    if checklist != [] : 
        cnt -= 1
print(cnt)
```

```python
# 7. 창고정리
import sys
#sys.stdin = open('input.txt','rt')
n = int(input())
list1 = list(map(int,input().split()))
m = int(input())
for _ in range(m) : 
    min_num = min(list1)
    max_num = max(list1)
    list1[list1.index(max_num)] -= 1
    list1[list1.index(min_num)] += 1
print(max(list1)-min(list1))
```

```python
# 8. 침몰하는 타이타닉
import sys
#sys.stdin = open('input.txt','rt')
n,m = map(int,input().split())
list1 = list(map(int,input().split()))
list4=[]
while list1 != [] :
    a = list1.pop(0)
    list2 = [a]
    list3 = [[idx,x] for idx,x in enumerate(list1) if x<=m-a]
    if list3 != [] : 
        list3.sort(key = lambda x : x[1],reverse=True)
        b = list1.pop(list3[0][0])
        list2.append(b)
    list4.append(list2)
print(len(list4))
```

```python
# 9. 증가수열 만들기
import sys
#sys.stdin = open('input.txt','rt')
n= int(input())
list1 = list(map(int,input().split()))
if list1[0] < list1[-1] : 
    lr = 'L'
else : 
    lr = 'R'
list2 = [list1.pop(list1.index(min(list1[0],list1[-1])))]
answer = [lr]
for _ in range(n) : 
    if max(list1[0],list1[-1]) < list2[-1] :
        break
    elif min(list1[0],list1[-1]) > list2[-1] : 
        if list1[0] < list1[-1] : 
            a = list1[0]
            lr = 'L'
        else : 
            a = list1[-1]
            lr = 'R'
    else : 
        if list1[0] > list1[-1] : 
            a = list1[0]
            lr = 'L'
        else : 
            a = list1[-1]
            lr = 'R'
    list2.append(list1.pop(list1.index(a)))
    answer.append(lr)
print(len(list2))
print(''.join(answer))
```
