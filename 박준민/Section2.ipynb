### 1. k번째 약수

작성하시오.

```python
# import sys
# sys.stdin=open('input.txt', 'rt')

# n, k = map(int, input().split())

n=8
k=4

cnt = 0
for i in range(1,n+1):
    if n % i == 0:
        cnt += 1
    if cnt == k:
        print(i)
        break
else:
    print(-1)
```

### 4. 대표값

N명의 학생 중 평균에 가장 가까운 학생은 몇 번째 학생인지 출력하는 프로그램을 작성하세
요.<br>

<br><br>
- 입력설명<br>
첫줄에 자연수 N(5<=N<=100)이 주어지고, 두 번째 줄에는 각 학생의 수학점수인 N개의 자연
수가 주어집니다. 학생의 번호는 앞에서부터 1로 시작해서 N까지이다.<br><br>
- 출력설명<br>
첫줄에 평균과 평균에 가장 가까운 학생의 번호를 출력한다.<br>
평균은 소수 첫째 자리에서 반올림합니다.<br><br>
- 입력예제 1<br>

45 73 66 87 92 67 75 79 75 80<br><br>
- 출력예제 1<br>

```python
import sys
sys.stdin=open('input.txt', 'rt')

n = map(int,input())
score = list(map(int, input().split()))
avg = round(sum(score)/len(score))

arrmin = score[0]
for idx, x in enumerate(score):
	if abs(avg-x) <= abs(avg-arrmin):
		if abs(avg-x) == abs(avg-arrmin):
			if x > arrmin:
				arrmin = x
				arridx = idx
		else:
			arrmin = x
			arridx = idx
print(avg, arridx + 1)


# 재상답안
#import sys
## sys.stdin = open('input.txt', 'rt')

#N = int(input())
#a = list(map(int, input().split()))
#avg = round(sum(a)/N)
#sol_dif = float('inf')
#sol_i = N
#sol = -float('inf')
#for i  in range(N) :
#    dif = abs(avg - a[i])
#    if (dif > sol_dif) :
#        continue
        
#    if (dif == sol_dif) & (a[i] <= sol) :
#        continue

#    sol = a[i]
#    sol_i = i
#    sol_dif = dif     
        
#print(int(avg), sol_i+1)
```

### 7. 소수(에라토스테네스 체)
자연수 N이 입력되면 1부터 N까지의 소수의 개수를 출력하는 프로그램을 작성하세요.<br>
만약 20이 입력되면 1부터 20까지의 소수는 2, 3, 5, 7, 11, 13, 17, 19로 총 8개입니다.<br>
제한시간은 1초입니다.<br><br>

- 입력설명<br>
첫 줄에 자연수의 개수 N(2<=N<=200,000)이 주어집니다.<br><br>

- 출력설명<br>
첫 줄에 소수의 개수를 출력합니다.<br><br>

- 입력예제 1<br>
20<br><br>

- 출력예제 1<br>
8

```python
import sys
sys.stdin=open('input.txt', 'rt')

n = int(input())

ans = 0
if n == 2:
	print(1)
else:
	for i in range(3, n+1, 2):
		cnt = 0
		for j in range(2, i+1):
			if i % j == 0:
				cnt += 1
		if cnt == 1:
			ans += 1
	else:
		print(ans+1)
        
# 제한시간 초과
```

프로그램을 작성하세요. 예를 들어 32를 뒤집으면 23이고, 23은 소수이다. 그러면 23을 출력

뒤집는 함수인 def reverse(x) 와 소수인지를 확인하는 함수 def isPrime(x)를 반드시 작성하
여 프로그래밍 한다.<br><br>

- 입력설명<br>
첫 줄에 자연수의 개수 N(3<=N<=100)이 주어지고, 그 다음 줄에 N개의 자연수가 주어진다.<br>

첫 줄에 뒤집은 소수를 출력합니다. 출력순서는 입력된 순서대로 출력합니다.<br><br>

- 입력예제 1<br>

- 출력예제 1<br>

```python
import sys
sys.stdin=open('input.txt', 'rt')

n = int(input())
num = list(map(int, input().split()))

# def reverse(x):
# 	a = str(x)[::-1]
# 	for i in range(len(a)):
# 		if a[0] == "0":
# 			a = a[1:]
# 		else:
# 			a = int(a)
# 			return a

def reverse(x):
    a = str(x)[::-1]
    return int(a)

def isPrime(x):
	if x != 1:
		for i in range(2, x):
			if x % i == 0 :
				break
		else:
			return x

for i in num:
	b = reverse(i)
	if isPrime(b) != None:
		print(isPrime(b), end=' ')
```

### 9. 주사위 게임
1에서부터 6까지의 눈을 가진 3개의 주사위를 던져서 다음과 같은 규칙에 따라 상금을 받는 게
임이 있다.<br><br>
규칙(1) 같은 눈이 3개가 나오면 10,000원+(같은 눈)*1,000원의 상금을 받게 된다.<br>
규칙(2) 같은 눈이 2개만 나오는 경우에는 1,000원+(같은 눈)*100원의 상금을 받게 된다.<br>
규칙(3) 모두 다른 눈이 나오는 경우에는 (그 중 가장 큰 눈)*100원의 상금을 받게 된다.<br><br>
N 명이 주사위 게임에 참여하였을 때, 가장 많은 상금을 받은 사람의 상금을 출력하는 프로그램
을 작성하시오<br><br>

- 입력설명<br>
첫째 줄에는 참여하는 사람 수 N(2<=N<=1,000)이 주어지고 그 다음 줄부터 N개의 줄에 사람
들이 주사위를 던진 3개의 눈이 빈칸을 사이에 두고 각각 주어진다.<br><br>

- 출력설명<br>
첫째 줄에 가장 많은 상금을 받은 사람의 상금을 출력한다.<br><br>

- 입력예제 1<br>
3<br>
3 3 6<br>
2 2 2<br>
6 2 5<br><br>

- 출력예제 1<br>
12000

```python
import sys
sys.stdin=open('input.txt', 'rt')

n = int(input())

m = []
for i in range(n):
	dice = list(map(int, input().split()))
	a = dice[0]
	b = dice[1]
	c = dice[2]
	if a == b == c:
		m.append(10000 + a*1000)
	elif a == b:
		m.append(1000 + a*100)
	elif b == c:
		m.append(1000 + b*100)
	elif a == c:
		m.append(1000 + a*100)
	else:
		m.append(max(dice)*100)

print(max(m))
```

### 10. 점수계산
여러 개의 OX 문제로 만들어진
시험에서 연속적으로 답을 맞히는 경우에는 가산점을 주기 위해서 다음과 같이 점수 계산을 하기
로 하였다.<br><br>
1번 문제가 맞는 경우에는 1점으로 계산한다.<br>
앞의 문제에 대해서는 답을 틀리다가
답이 맞는 처음 문제는 1점으로 계산한다.<br>
또한, 연속으로 문제의 답이 맞는 경우에서 두 번째
문제는 2점, 세 번째 문제는 3점, ..., K번째 문제는 K점으로 계산한다.<br>
틀린 문제는 0점으로 계산한다.<br><br>
시험문제의 채점 결과가 주어졌을 때, 총 점수를 계산하는 프로그램을 작성하시오.<br><br>

- 입력설명<br>
첫째 줄에 문제의 개수 N (1 ≤ N ≤ 100)이 주어진다. <br>
둘째 줄에는 N개 문제의 채점 결과를 나타내는 0 혹은 1이 빈 칸을 사이에 두고 주어진다.<br>
0은 문제의 답이 틀린 경우이고, 1은 문제의답이 맞는 경우이다.<br><br>

- 출력설명<br>
첫째 줄에 입력에서 주어진 채점 결과에 대하여 가산점을 고려한 총 점수를 출력한다.<br><br>

- 입력예제 1<br>
10<br>
1 0 1 1 1 0 0 1 1 0<br><br>

- 출력예제 1<br>
10

```python
import sys
sys.stdin=open('input.txt', 'rt')

n = int(input())
num = list(map(int, input().split()))

p = []
for idx, i in enumerate(num):
	if idx == 0:
		p.append(i)
	else:
		a = p[-1]*i +i
		p.append(a)
print(sum(p))
```
